```python
import numpy as np

def get_beans(counts):
    xs = np.random.rand(counts)
    xs = np.sort(xs)
    ys = [1.2*x+np.random.rand()/10 for x in xs]
    return xs,ys
```

```python
xs,ys = get_beans(100)
print(xs)
print(ys)
```

```python
from matplotlib import pyplot as plt

plt.title("Size-Toxicity Function", fontsize=12)
plt.xlabel("Bean Size")
plt.ylabel("Toxicity")

plt.scatter(xs, ys)
plt.show()
```

```python
w = 1.2827209903409331#0.1

y_pre = w * xs
plt.scatter(xs, ys)
plt.plot(xs, y_pre)
plt.show()
```

```python
es = (ys - y_pre)**2
sum_e = np.sum(es)
sum_e = 1/len(xs)*sum_e
print(sum_e)
```

```python
ws = np.arange(0,3,0.1)
print(ws)
```

```python
es = []
m = len(xs)
for w in ws:
    y_pre = w * xs
    e = (1/m) * np.sum((ys - y_pre)**2)
    es.append(e)
    
plt.title("cost function", fontsize=12)
plt.xlabel("w")
plt.ylabel("e")
plt.plot(ws, es)
plt.show()

w_min = np.sum(xs * ys) / np.sum(xs * xs)
print(w_min)
```

```python
import random
import numpy as np

def get_beans( size ):
    w = 1.0
    xs = []
    ys = []
    for i in range(size):
        x = i
        xs.append(x / 100)
        ys.append((w * x + random.random() * 10) / 100) 
    return np.array(xs),np.array(ys)
```

```python
print(get_beans(5))
```

```python
xs,ys = get_beans(100)
print(xs)
print(ys)
```

```python
from matplotlib import pyplot as plt

plt.title("Size-Toxicity Function", fontsize=12)
plt.xlabel("Bean Size")
plt.ylabel("Toxicity")

plt.scatter(xs, ys)
plt.show()
```

```python
w = 0.5
for x in range(100):
    for i in range(len(xs)):
        x = xs[i]
        y = ys[i]
        y_pre = w * x
        e = y - y_pre
        alpha = 0.05 # 学习率
        w = w + alpha * e * x
        
y_pre1 = w * xs
plt.scatter(xs, ys)
plt.plot(xs, y_pre1)
plt.show()
```

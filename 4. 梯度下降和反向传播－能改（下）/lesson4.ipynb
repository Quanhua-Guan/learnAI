```python
import dataset
from matplotlib import pyplot as plt
import numpy as np

m = 100
xs,ys = dataset.get_beans(100)

plt.title("Size-Toxicity Function", fontsize=12)
plt.xlabel("Bean Size")
plt.ylabel("Toxicity")
plt.xlim(0,1)
plt.ylim(0,1.5)

plt.scatter(xs, ys)

w = 0.1
b = 0.1
y_pre = w * xs + b

plt.plot(xs, y_pre)

plt.show()
```

```python
#lesson3 
# e = x^2*w^2 + (-2xy)*w +
w = 0.1
b = 0.1
size = len(xs)
for _ in range(100):
    for i in range(size):
        x = xs[i]
        y = ys[i]

        #a=x^2
        #b=-2*x*y
        #c=y^2
        # k=2*a*w+b
        dk = 2*(x**2)*w + 2*x*b + (-2*x*y)
        db = 2*b + 2*x*w -2*y

        # w
        alpha = 0.01
        w = w - alpha*dk # w.
        b = b - alpha*db
        
    plt.clf()
    plt.scatter(xs,ys)
    y_pre = w*xs+b
    plt.xlim(0,1)
    plt.ylim(0,1.2)
    plt.plot(xs,y_pre)
    plt.pause(0.01)

# 
plt.scatter(xs, ys)
y_pre = w * xs + b
plt.plot(xs, y_pre)
plt.show()
        
    
```
